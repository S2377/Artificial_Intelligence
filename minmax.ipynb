3. Computer's Turn.

This will happen repeatedly.

```python
def is_terminal(state):
    
    ### Horizontal case ####
    if state[0] == state[1] == state[2] == 'X':
        return True, 1
    elif state[3] == state[4] == state[5] == 'X':
        return True, 1
    elif state[6] == state[7] == state[8] == 'X':
        return True, 1
    
    #### Vertical case #####
    elif state[0] == state[3] == state[6] == 'X':
        return True, 1
    elif state[1] == state[4] == state[7] == 'X':
        return True, 1
    elif state[2] == state[5] == state[8] == 'X':
        return True, 1
    
    ### Diagonal  Case #####
    elif state[0] == state[4] == state[8] == 'X':
        return True, 1
    elif state[2] == state[4] == state[6] == 'X':
        return True, 1
    
    ##### Horizontal case ######
    elif state[0] == state[1] == state[2] == 'O':
        return True, -1
    elif state[3] == state[4] == state[5] == 'O':
        return True, -1
    elif state[6] == state[7] == state[8] == 'O':
        return True, -1
    
    ###### Vertical case #####
    elif state[0] == state[3] == state[6] == 'O':
        return True, -1
    elif state[1] == state[4] == state[7] == 'O':
        return True, -1
    elif state[2] == state[5] == state[8] == 'O':
        return True, -1
    
    ###### Diagonal case #######
    elif state[0] == state[4] == state[8] == 'O':
        return True, -1
    elif state[2] == state[4] == state[6] == 'O':
        return True, -1
    elif state.count('-') == 0:
        return True, 0
    else:
        return False, 0
```

```python
def player_move(state):
    num = state.count('-')
    if num % 2 == 0:
        return 'O'
    else:
        return 'X'
```

```python
def list_to_string(lst):
    string=''
    for i in lst:
        string+=i
    return string
```

```python
# Initialize the game tree, queue, terminal states, parent-child relationships, and scores
tree = {}
queue = [['-']*9]
terminal = []
child_parent = {}
Scores = {}

# Function to create the game tree using a breadth-first search
def create_tree(state = ['-']*9):
    global tree, queue
    queue.pop(0)
    # Convert the state to a string for dictionary indexing
    string_of_state = list_to_string(state)
    tree[string_of_state] = []
    # Determine the current player's turn
    turn = player_move(state)
     # Check if the current state is not terminal
    if not is_terminal(state)[0]:
        
        for i in range(9):
            state_copy = state.copy()
            # Check if the move is valid
            if state[i] == '-':
                 # Make the move and update the state
                state_copy[i] = turn
                state_copy_string = list_to_string(state_copy)
                tree[string_of_state].append(state_copy_string)
                queue.append(state_copy)
                 # Update parent-child relationships
                if state_copy_string not in child_parent.keys():
                    child_parent[state_copy_string] = []
                score = is_terminal(state_copy_string)[1]
                Scores[state_copy_string]=[score, turn]
                if string_of_state not in child_parent[state_copy_string]:
                    child_parent[state_copy_string].append(string_of_state)
                
                # Recursively create the tree for the new state
                create_tree(state_copy)
    else:
        if string_of_state not in terminal:
            terminal.append(string_of_state)
        

create_tree()
child_parent['---------']=[None]
Scores['---------']=[0, 'O']
```

```python
new_terminal = []

for i in terminal:
    if i not in new_terminal:
        new_terminal.append(i)
        
terminal = new_terminal
```

```python
# Function to recursively update scores based on minimax algorithm
def update_helper(parents):
    # Continue updating until reaching the root (parent is None)
    while parents[0] is not None:
        for j in parents:
            score_list=[]
            children = tree[j]
            # Collect scores of children nodes
            for k in children:
                score_list.append(Scores[k][0])
                
            # Update the score of the current node based on player's turn
            if Scores[j][1] == 'X':
                Scores[j][0] = min(score_list)
            else:
                Scores[j][0] = max(score_list)
                
            # Update parents for the next iteration
            parents = child_parent[j]
            # Recursively update scores for parent nodes
            update_helper(parents)
            
            
# Function to update scores for all terminal nodes           
def update_values():
    # Iterate over all terminal nodes
    for i in terminal:
        parents = child_parent[i]
        # Update scores for each terminal node and its ancestors
        update_helper(parents)
        
update_values()
        
```

```python
# Function to find the position to make a move based on the action and the current state
def position_for_move(state, action):
    for i in range(9):
        # Find the position where the state has '-' (empty) and the action has a different value
        if state[i] == '-' and action[i] != '-':
            return i

# Function to determine the best move for the current state
def best_move(state, player):
    # Convert the state to a string for dictionary indexing
    state = list_to_string(state)
    # Get the children of the current state
    children = tree[state]
    # Dictionary to store scores and corresponding moves
    score_dict = {}
    # Iterate over children and store their scores and moves in 'score_lst'
    for i in children:
        score_dict[Scores[i]] = i

    score = min(score_dict.keys())
    # Return the position to make the best move
    return position_for_move(state, score_dict[score])
```

```python
def user_move_index():
    # taking input from the user
    row = int(input("Enter the row for move:- "))
    column = int(input("Enter the column for the move:- "))
    if row == 0:
        index = column -1
    elif row == 1:
        index = 2 * row + column
    else:
        index = 2 * row + column + 1
    return index
    
```

```python
# game board
def display(state):
    print('-------------------')
    print('| ',state[0],' | ',state[1],' | ',state[2],' |')
    print('-------------------')
    print('| ',state[3],' | ',state[4],' | ',state[5],' |')
    print('-------------------')
    print('| ',state[6],' | ',state[7],' | ',state[8],' |')
    print('-------------------')
```

```python
def play_game():
    state = ['-']*9
    display(state)
    player = 0
    while not is_terminal(state)[0]:
        if player % 2 ==0:
            index = user_move_index()
            state[index] = 'X'
        else:
            index = best_move(state)
            state[index]='O'
        display(state)
        player += 1
        
        terminal = is_terminal(state)
        if terminal[1] == -1:
            return 'Computer won'
        elif terminal[1] == 1:
            return 'You won'
        elif terminal[0] and terminal[1] == 0:
            return 'Match draw'
        
```
